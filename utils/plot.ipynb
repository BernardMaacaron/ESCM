```python
import matplotlib.pyplot as plt

# Define the grid size
grid_size = 11

# Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111)

# Plot neurons as blue squares in an 11x11 grid
for x in range(grid_size):
    for y in range(grid_size):
        ax.add_patch(plt.Rectangle((x - 0.5, y - 0.5), 1, 1, edgecolor='black', facecolor='blue', visible=True))

# Find the central neuron
central_x, central_y = grid_size // 2, grid_size // 2

# Highlight the central neuron in green
ax.add_patch(plt.Rectangle((central_x - 0.5, central_y - 0.5), 1, 1, edgecolor='black', facecolor='green'))

# Set the aspect ratio to be equal
ax.set_aspect('equal')
ax.set_xlim(-0.5, grid_size - 0.5)
ax.set_ylim(-0.5, grid_size - 0.5)
# Set labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Neuron Neighborhood')

# Show the plot
plt.show()
```
